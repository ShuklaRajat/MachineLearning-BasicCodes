```python
import pandas as pd
def specific_hypothesis(data):
    (nr,nc)=data.shape
    hyp=['0']*(nc-1)
    spec_list=[]
    first_no=0
    for row in range (nr):
        if data.iloc[row,-1]=="yes":
            for col in range (nc-1):
                if hyp[col]=='0':
                    hyp[col]=data.iloc[row,col]
                elif hyp[col]!=data.iloc[row,col]:
                    hyp[col]='?'
        elif first_no==0:
            first_no=row
        spec_list.append(list(hyp))
        
    return spec_list,first_no

data= pd.read_csv('ml.csv')
print(data)
spec_hyp,first_no = specific_hypothesis(data)
print("first_no", first_no)
def generic_hypothesis(data,first_no):
    (nr,nc)=data.shape
    gen_hyp_list=[]
    elem_list=[]
    for row in range(nr):
        if row<first_no:
            gen_hyp_list.append(['?']*(nc-1))
        if data.iloc[row,-1]=="no":
           for col in range(nc-1):
                if(data.iloc[row,col] != spec_hyp[row][col] and spec_hyp[row][col] !='?'):
                   new_gen_hyp=['?']*(nc-1)
                   new_gen_hyp[col]=spec_hyp[row][col]
                   elem_list.append(new_gen_hyp)
           
           gen_hyp_list.append(elem_list)
           elem_list=[]
        elif row>first_no and data.iloc[row,-1]=="yes":
            for col in range(nc-1):
                if(spec_hyp[row-1][col]!=spec_hyp[row][col]):
                    change_att=spec_hyp[row-1][col]
            pos_gen_list=gen_hyp_list[row-1].copy()
            for rows in pos_gen_list:
                if change_att in rows:
                    pos_gen_list.remove(rows)
            gen_hyp_list.append  (pos_gen_list)
            
    return gen_hyp_list

gen = generic_hypothesis(data,first_no)
print("\nSpecific\n")
print(*spec_hyp ,sep="\n")
print("\nGeneral\n")
print(*gen ,sep="\n")
```
